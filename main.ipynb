```python
import model
import detect
import eyes
import os
```

```python
WIDTH, HEIGHT = 28, 10 
dim = WIDTH * HEIGHT
```

```python
def title():
    os.system('cls')

    print("\t\t\t\t\t***************************************************")
    print("\t\t\t\t\t  ***** Contoh Abstrak - Sistem Emosi Wajah *****  ")
    print("\t\t\t\t\t***************************************************")


def mainMenu():
    title()
    print()
    print(9*"*", "Pilihan Menu", 9*"*")
    print("[1] training model")
    print("[2] emotion detection")
    print("[3] Eyes Detection")
    print("[4] Quit")

    while True:
        try:
            choice = int(input("Pilihan Anda: "))

            if choice == 1:
                modelDetect()
                break
            elif choice == 2:
                emotionDetect()
                break
            elif choice == 3:
                eyesDetect()
                break
            elif choice == 4:
                print("Thank You")
                break
            else:
                print("Maaf, pilihan yang anda masukkan tidak tersedia")
                mainMenu()
        except ValueError:
            print("Maaf, pilihan yang anda masukkan tidak tersedia. SIlahkan coba kembali")
    exit
```

```python
def modelDetect(): 
    model.model() 
    key = input("Enter any key to return main menu") 
    mainMenu()

def emotionDetect(): 
    detect.detect() 
    key = input("Enter any key to return main menu") 
    mainMenu()

def eyesDetect(): 
    eyes.eyes() 
    key = input("Enter any key to return main menu") 
    mainMenu()

mainMenu()
```
